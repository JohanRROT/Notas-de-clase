```python
#comporbacion de una matriz ortogonal, para ello hay que comprobar que sus columnas y filas son 
#ortogonales entre si
matriz=np.array([[1,0,0],[0,1,0],[0,0,1]])
print(matriz)
```

```python
#comprobacion de ortogonalidad de las columnas(vectores) 
print(matriz[:,0].dot(matriz[:,1]))
print(matriz[:,0].dot(matriz[:,2]))
print(matriz[:,1].dot(matriz[:,2]))
print(matriz[:,1])#eso signica todos elementos de la columna
```

```python
#si son ortogonales ahora solo resta verificar si son ortonormales
print(np.linalg.norm(matriz[:,0]))
print(np.linalg.norm(matriz[:,1]))
print(np.linalg.norm(matriz[:,2]))
```

```python
#se hace el mismo proceso con las filas 
print(matriz[0,:].dot(matriz[1,:]))
print(matriz[0,:].dot(matriz[2,:]))
print(matriz[1,:].dot(matriz[2,:]))
print(matriz[1,:])#eso signica todos elementos de la fila
```

```python
#por ultimo comprobamos si tambien son ortonormales
print(np.linalg.norm(matriz[0,:]))
print(np.linalg.norm(matriz[1,:]))
print(np.linalg.norm(matriz[2,:]))
```

```python
#podemos genrar una matriz ortogonal de la siguiente forma
A=np.array([[np.cos(100),-np.sin(100)],
           [np.sin(100),np.cos(100)]])
print(A)
```

```python
#para comprobarlo podemos revisar su norma, el valor queda en 0.999.. por problemas computacionales 
print(np.linalg.norm(A[0,:]))
print(np.linalg.norm(A[1,:]))
print(np.linalg.norm(A[:,0]))
print(np.linalg.norm(A[:,1]))
```

```python
#ahora comprobamos sin son ortogonales 
print(A[0,:].dot(A[1,:]))
print(A[:,0].dot(A[:,1]))
```

```python
A_t=A.T #podemos comprobar multiplicando la matriz por su transpuesta y esto regresara la matriz identidad
print(A_t.dot(A))
print(A.dot(A_t))
```
