```python
%matplotlib notebook 
#esto nos permite hacer graficos que sean interactivos 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-6,6)

y_1= 3*x+5
y_2= -x+3
y_3= 2*x+1
```

```python
plt.figure() #este sistema que estamos graficando es un sistema sobredeterminado, tiene mas ecuaciones que variables, si revisamos
plt.plot(x, y_1) #el grafico se evidencia que no hay un punto en el que las tres lineas confluyan, por ende no tiene solucion 
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```
